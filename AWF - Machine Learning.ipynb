```
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
# from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import gc
import time

from sklearn.model_selection import GridSearchCV
```

```
with open('data/data.npy', 'rb') as f:
    data_npy = np.load(f, allow_pickle=True)
    
print('Data Loading Complete')
    
with open('data/labels.npy', 'rb') as f:
    labels_npy = np.load(f, allow_pickle=True)
    
print('Label Loading Complete')  
```

```
le = preprocessing.LabelEncoder()
le.fit(labels_npy)
labels_cat = le.transform(labels_npy)
print('Label Encoding Complete')
```

```
X_train, X_test, y_train, y_test = train_test_split(data_npy, labels_cat, test_size=0.2, random_state=105)
print('Split Complete')
```

```
result_df = pd.DataFrame()
```

```
del data_npy, labels_npy, labels_cat, le
gc.collect()
```

```
model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=105, verbose=1) 

gc.collect()

print('Training RF..')
start_time = time.time()
model.fit(X_train, y_train)
train_time = time.time() - start_time
print('Training Complete')

gc.collect()
# y_pred = model.predict(X_test)
start_time = time.time()
acc = model.score(X_test, y_test)
inf_time = time.time() - start_time
new_row = {'Model': 'RF', 
           'Train Time': train_time, 
           'Inference Time': inf_time,
           'Accuracy': acc}
result_df = result_df.append(new_row, ignore_index=True)
print('Accuracy (RF):', acc)
```

```
model = GaussianNB()

gc.collect()

print('Training Gaussian NB..')
start_time = time.time()
model.fit(X_train, y_train)
train_time = time.time() - start_time
print('Training Complete')

gc.collect()
# y_pred = model.predict(X_test)
start_time = time.time()
acc = model.score(X_test, y_test)
inf_time = time.time() - start_time
new_row = {'Model': 'GaussianNB', 
           'Train Time': train_time, 
           'Inference Time': inf_time,
           'Accuracy': acc}
result_df = result_df.append(new_row, ignore_index=True)
print('Accuracy (GaussianNB):', acc)
```

```
model = AdaBoostClassifier(random_state=105)

gc.collect()

print('Training Ada..')
start_time = time.time()
model.fit(X_train, y_train)
train_time = time.time() - start_time
print('Training Complete')

gc.collect()
# y_pred = model.predict(X_test)
start_time = time.time()
acc = model.score(X_test, y_test)
inf_time = time.time() - start_time

new_row = {'Model': 'Ada', 
           'Train Time': train_time, 
           'Inference Time': inf_time,
           'Accuracy': acc}
result_df = result_df.append(new_row, ignore_index=True)
print('Accuracy (Ada):', model.score(X_test, y_test))
```

```
model = KNeighborsClassifier(n_jobs=-1) 

gc.collect()

print('Training KNN..')
start_time = time.time()
model.fit(X_train, y_train)
train_time = time.time() - start_time
print('Training Complete')

gc.collect()
# y_pred = model.predict(X_test)
start_time = time.time()
acc = model.score(X_test, y_test)
inf_time = time.time() - start_time
new_row = {'Model': 'KNN', 
           'Train Time': train_time, 
           'Inference Time': inf_time,
           'Accuracy': acc}
result_df = result_df.append(new_row, ignore_index=True)
print('Accuracy (KNN):', acc)
```

```
rf_result = pd.DataFrame()
```

```
estimators = [50, 75, 100, 125]


for est in estimators:

  model = RandomForestClassifier(n_estimators=est, n_jobs=-1, random_state=105, verbose=1) 

  gc.collect()

  print(f'Training RF with {est} estimators')
  start_time = time.time()
  model.fit(X_train, y_train)
  train_time = time.time() - start_time
  print('Training Complete')

  gc.collect()
  # y_pred = model.predict(X_test)
  start_time = time.time()
  acc = model.score(X_test, y_test)
  inf_time = time.time() - start_time
  new_row = {'Model': 'RF-'+str(est), 
            'Train Time': train_time, 
            'Inference Time': inf_time,
            'Accuracy': acc}
  rf_result = rf_result.append(new_row, ignore_index=True)
  del model
  gc.collect()
  print(f'Accuracy (RF) with {est} estimators: {acc}')
```
